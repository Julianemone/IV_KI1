## Part B: Correlations 


```python
IHME_data = pd.read_csv('IHME_data.csv')
covid_data = pd.read_csv('owid_data.csv')
```

```python
covid_data['date'] = pd.to_datetime(covid_data['date'])
covid_data = covid_data[covid_data['date'].dt.year.isin([2019, 2020, 2021])]  # Example for 2019 and 2021

# List columns to discard after aggregation
columns_to_discard = ['new_cases', 'new_deaths', 'new_cases_smoothed', 'new_deaths_smoothed', 'new_cases_per_million', 'new_cases_per_million', 'new_cases_smoothed_per_million']

# Aggregate the dataset by country
covid_agg = covid_data.groupby('location').agg({
    'new_cases': 'sum',
    'new_deaths': 'sum',
}).reset_index()

# Drop the irrelevant columns
covid_agg = covid_agg.drop(columns=columns_to_discard, errors='ignore')


# Join the datasets on the country key
merged_data = pd.merge(IHME_data, covid_agg, on='location', how='inner')

# Save or use the merged dataset as needed
merged_data.to_csv('merged_dataset.csv', index=False)
```
